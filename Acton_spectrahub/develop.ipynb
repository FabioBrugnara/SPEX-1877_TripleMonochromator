```python
import ACTON_SpectraHub_driver as SpectraHub
import matplotlib.pyplot as plt
import time
```

```python
detector = SpectraHub.connect_detector()
```

```python
detector.read()
```

```python
detector.set_itime(1000)
```

```python
# library for empty cell output
from IPython.display import clear_output
```

```python
counts = []
detector.empty_buffer()
fig, ax = plt.subplots()
try :
    while True:
        clear_output(wait=True)
        counts.append(detector.read(verbose=False))
        plt.plot(counts)
        plt.ylabel('Current [A]')
        plt.show()
        time.sleep(0.1)
except KeyboardInterrupt:
    detector.empty_buffer()

plt.plot(counts)
```
